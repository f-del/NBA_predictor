```python
def create_player_info(data):
    """Create and transform the player information DataFrame."""
    df = pd.DataFrame(data)
    if "Position" in df.columns:
        df = transform_position_to_array(df, "Position")
    if "Weight" in df.columns:
        df = split_weight_units(df, "Weight")
    if "Height" in df.columns:
        df = convert_height_to_meters(df, "Height")
    df = convert_dates_to_datetime(df, ["Birthdate", "NBA Debut"])
    return df


def create_player_stats(stats_data):
    """Create and calculate metrics for the player stats DataFrame."""
    stats_df = pd.DataFrame(stats_data)
    stats_df["PTS/G"] = stats_df["PTS"] / stats_df["G"].replace(0, np.nan)
    stats_df["TRB/G"] = stats_df["TRB"] / stats_df["G"].replace(0, np.nan)
    stats_df["AST/G"] = stats_df["AST"] / stats_df["G"].replace(0, np.nan)
    stats_df["BLK/G"] = stats_df["BLK"] / stats_df["G"].replace(0, np.nan)
    stats_df["MP/G"] = stats_df["MP"] / stats_df["G"].replace(0, np.nan)
    return stats_df
```

```python
import pandas as pd
from tabulate import tabulate


# Creating a DataFrame for player stats
stats_df = pd.DataFrame(stats_data)

# Dropping unwanted columns from a copy of stats_df
filtered_stats_df = stats_df.drop(columns=[
    "Awards", "Trp-Dbl", "FG%", "3P%", "2P%", "eFG%", "FT%", "Lg"
])

# Displaying the filtered stats DataFrame as a table
print("\nFiltered Player Stats:")
print(tabulate(filtered_stats_df[:2], headers='keys', tablefmt='grid'))
```

```python


def calculate_improvements(filtered_seasons):
    """Calculate statistical improvements between seasons."""
    return {
        "PTS/G": filtered_seasons["PTS/G"].iloc[1] - filtered_seasons["PTS/G"].iloc[0],
        "BLK/G": filtered_seasons["BLK/G"].iloc[1] - filtered_seasons["BLK/G"].iloc[0],
        "3P%": filtered_seasons["3P%"].iloc[1] - filtered_seasons["3P%"].iloc[0],
    }



def plot_enhanced_charts(filtered_seasons, improvement_stats):
    """Plot enhanced charts and improvements."""

    plt.style.use('default')  # Reset to default style with white background

    def plot_single_chart(ax, x, y, title, xlabel, ylabel, color, ylim_factor=1.2, annotate=True, format_value="{:.2f}"):
        """Helper function to plot a single chart."""
        ax.bar(x, y, color=color)
        ax.set_title(title)
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.grid(axis='y', linestyle='--', alpha=0.7)
        ax.set_ylim(0, max(y) * ylim_factor)
        if annotate:
            for i, val in enumerate(y):
                ax.text(i, val + (max(y) * 0.02), format_value.format(val), ha='center', fontsize=10)

    fig, axes = plt.subplots(1, 3, figsize=(18, 6))

    # Points Per Game
    plot_single_chart(
        axes[0],
        filtered_seasons["Season"],
        filtered_seasons["PTS/G"],
        "Points Per Game",
        "Season",
        "PTS/G",
        color="blue"
    )

    # Blocks Per Game
    plot_single_chart(
        axes[1],
        filtered_seasons["Season"],
        filtered_seasons["BLK/G"],
        "Blocks Per Game",
        "Season",
        "BLK/G",
        color="purple"
    )

    # 3-Point Percentage
    plot_single_chart(
        axes[2],
        filtered_seasons["Season"],
        filtered_seasons["3P%"],
        "3-Point Percentage",
        "Season",
        "3P%",
        color="teal",
        format_value="{:.3f}"
    )

    plt.tight_layout()
    plt.show()

    # Improvement Bar Chart
    plt.figure(figsize=(8, 6))
    plt.bar(improvement_stats.keys(), improvement_stats.values(), color=["blue", "purple", "teal"])
    plt.title("Statistical Improvements (2023-24 to 2024-25)")
    plt.ylabel("Improvement")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    for i, (key, val) in enumerate(improvement_stats.items()):
        plt.text(i, val + 0.01, f"{val:.2f}", ha='center', fontsize=10)
    plt.show()

```

```python

player_info_df = create_player_info(data)
print(tabulate(player_info_df, headers='keys', tablefmt='grid'))



player_stats_df = create_player_stats(stats_data)
filtered_seasons = player_stats_df[player_stats_df["Season"] != "2 Yrs"]
improvements = calculate_improvements(filtered_seasons)

plot_enhanced_charts(filtered_seasons, improvements)
```
